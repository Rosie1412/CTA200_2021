```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
d,f,m,α,β=sympy.symbols('d,f,m,α,β')
λ=(d/m)*(sin(α)+sin(β))
D=1/(diff(λ,β))
D
```

```python
x=np.arange(0,3.14,.1)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel('Diffraction angle β (rad)', fontsize=14)
plt.ylabel('Spectral resolution Δλ (arb)',fontsize=14)
plt.title('Conceptual relationship between β and Δλ',fontsize=14)
```

```python
x=np.arange(.1,2,.01)
y=(1/x)
plt.plot(x,y)
plt.xlabel('Focal length f (arb)',fontsize=14)
plt.ylabel('Spectral resolution Δλ (arb)',fontsize=14)
plt.title('Conceptual relationship between f and Δλ',fontsize=14)
```
