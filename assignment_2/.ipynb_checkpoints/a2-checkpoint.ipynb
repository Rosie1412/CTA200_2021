```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

## Question 1

```python
#taylor approximation used to define numerical derivative
def deriv(f,x0,h):
    dx_fx0 = (f(x0+h)-f(x0-h))/(2*h)
    return(dx_fx0)
```

```python
#setting range of h values
h_arr=np.arange(.1,1,.001)

#calculating numerical and analytical derivative at .1 for each h value
dnum=[]
dana=[]
for i in h_arr:
    dnum.append(deriv(np.sin,.1,i))
    dana.append(np.cos(i))
dnum=np.array(dnum)
dana=np.array(dana)

#calculating error between numerical and analytical values
err=np.abs((dnum-dana)/dana)
```

```python
#plotting
plt.plot(h_arr,err)
plt.title('Stepsize vs Error',fontsize=14)
plt.xlabel('Stepsize',fontsize=14)
plt.ylabel('Error in numerical derivative',fontsize=14)
plt.show()
plt.savefig('fig1')
```

## Question 2


```python
#setting axes and variables
x=np.linspace(-2,2,1000)
y=np.linspace(-2,2,1000)
X,Y=np.meshgrid(x,y)
z=0
C=X+Y*1j

#passes through 100 iterations of the sequence, saving values that diverge into array
for i in range(100):
    z=z**2+C
    divergent=np.isnan(z)

#plotting
fig,ax=plt.subplots(figsize=(8,6))
ax.imshow(divergent,extent=[-2,2,-2,2],cmap='viridis',vmin=0,vmax=1)
plt.xlabel("x",fontsize=14)
plt.ylabel("y*i",fontsize=14)
plt.title("Points of convergence (purple) and divergence (yellow)",fontsize=12)
plt.show()
plt.savefig('fig21')
```

```python
#passes through up to 100 iterations to save each point of divergence
z=0
iteration = np.zeros_like(C).astype(int)
for i in range(100):
    z=z**2+C
    iteration+=np.isfinite(z)
    
#for cases where divergence does not occur within 100 iterations, value is 
#set to -20 to create color contrast
iteration=np.where(iteration==100,-20,iteration)

#plotting
plt.figure(figsize=(8,6))
plt.contourf(X,Y,iteration,levels=50)
plt.colorbar(label='iteration divergence number')
plt.xlabel('x',fontsize=14)
plt.ylabel('y*i',fontsize=14)
plt.title('Range of divergence iterations',fontsize=14)
plt.show()
plt.savefig('fig22')
```

## Question 3

```python
#defining ODE functions of SIR model
def SIR_mod(y,t,beta,gamma):
    S,I,R=y
    
    dS_dt=-beta*S*I*(1/1000)
    dI_dt=beta*S*I*(1/1000)-gamma*I
    dR_dt=gamma*I
    
    return(dS_dt,dI_dt,dR_dt)
```

```python
#setting initial conditions
S0=999
I0=1
R0=0
t=np.linspace(0,200,10000)
```

```python
#odeint used to integrate functions based on values of beta and gamma
def SIR_solution(beta,gamma):
    
    solution=odeint(SIR_mod,[S0,I0,R0],t,args=(beta,gamma))
    solution=np.array(solution)

    #plotting, with title autogenerated based on (beta,gamma) values
    params=(beta,gamma)
    plt.plot(t,solution[:,0],label="succeptible")
    plt.plot(t,solution[:,1],label="infected")
    plt.plot(t,solution[:,2],label="recovered")
    plt.ylabel('number of people',fontsize=14)
    plt.xlabel('time (arb)',fontsize=14)
    plt.title(f"SIR model: (beta,gamma)={params}",fontsize=14)
    plt.legend()
    plt.show()
```

```python
SIR_solution(1,.01)
plt.savefig('fig31')
```

```python
SIR_solution(1,.1)
plt.savefig('fig32')
```

```python
SIR_solution(1,1)
plt.savefig('fig33')
```
